```python
import pandas as pd 
import math
from geopy.distance import geodesic
import csv
from pandas import *
import numpy as np

df = pd.read_csv(r"C:\Users\yaoz\OneDrive - BAT\Desktop\Complete Distinct Address.csv")
df
```

```python
### This code stores 257 Expressway exits as a JSON? Dictionary? and then for each of the 9500+ distinct addresses, computes the closest exit with the distance and exit name
import pandas as pd 
import math
from geopy.distance import geodesic

def function(latitude,longitude):
    lst = []    
    try:
        address = (latitude,longitude)
    
        for index,row in df2.iterrows():
            exit = (row['results.LATITUDE'],row['results.LONGITUDE'])
            lst.append(geodesic(address,exit).km)
    
    
        min_value = min(lst)
        min_index = lst.index(min_value) 
    
        with open('output.csv', 'a', encoding='UTF8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([min_index, min_value]) 
            
    except ValueError:
        with open('output.csv', 'a', encoding='UTF8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow("zero")
        
        
```

```python
for index, row in df.iterrows():
    i = 0
    try:
        function(row['latitude'], row['longitude']) 
    except IndexError:
        with open('output.csv', 'a', encoding='UTF8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow("error occurred")
        continue
    i+=1 
    if i % 1000 == 0:
        print(i)
```
