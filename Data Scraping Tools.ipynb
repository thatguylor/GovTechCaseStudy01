```python
#This segment of the code will load a pre-built table cleaned in Power BI based on Data.gov Resale HDB Prices. 
### Motive is to use OneMap API to get both postal code and LAT and LONG for these addresses and write them back to the same csv file.###
import pandas as pd 
import csv
from pandas import *
import numpy as np
df = pd.read_csv(r"C:\Users\yaoz\OneDrive - BAT\Desktop\Distinct Addresses.csv")
df

```

```python
#Pseudo code -> For each row in dataframe, iterate over columns street_name and block 
#Using result = Client.Search(street_name + block)
#latitude = result.get('results')[0].get('LATITUDE'), longitude = result.get('results')[0].get('LONGITUDE')
#write latitude val to column LAT and write longitude val to column LONG
#Next row

def iterativefunction(street_name,block):
    result = Client.search(street_name + " "+ block)
    latitude = result.get('results')[0].get('LATITUDE') 
    longitude = result.get('results')[0].get('LONGITUDE')
    with open('results.csv', 'a', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([latitude,longitude])
    f.close() 

i = 0    
    
for index, row in df.iterrows():
    try:
        iterativefunction(row['street_name'], row['block']) 
    except IndexError:
        with open('results.csv', 'a', encoding='UTF8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow("error occurred")
        continue
    i+=1 
    if i % 1000 == 0:
        print(i)

    
```
